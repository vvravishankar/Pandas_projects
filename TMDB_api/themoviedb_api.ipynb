```python
import pandas as pd
import requests
import json
```

```python
# Making the Api request to get the data. api respon in json format
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=760c6998a263efbcc59e58a7a27a08b4&langauage=en=US&page=1')
response.json()
```

```python
# selecting the required key
response.json()['results']
```

```python
#converting the json to DF
df = pd.DataFrame(response.json()['results'])
df
```

```python
# selecting the required columns
df1 = df[['id','original_title','overview','release_date','popularity','vote_average','vote_count']]
df1
```

```python
# Initialize an empty list to store DataFrames
dfs = []

# itrate through each
for i in range(1, 2):  # Adjust the range as needed
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=760c6998a263efbcc59e58a7a27a08b4&language=en-US&page={}'.format(i))
    temp_df = pd.DataFrame(response.json()['results'])[['id', 'original_title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']]
    dfs.append(temp_df)

# Concatenate all DataFrames in the list into one DataFrame
Final_df = pd.concat(dfs, ignore_index=True)

# Output the final DataFrame
print(Final_df)
```
