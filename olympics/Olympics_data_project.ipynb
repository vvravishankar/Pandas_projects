```python
# Here the 01 02 03 are the Gold,Silver,Bronze medals counts lets rename the,
col_names = {  'Unnamed: 0': 'Country',
               '? Summer': 'Summer Olympics',
               '01 !': 'Gold',
               '02 !': 'Silver',
               '03 !': 'Bronze',
               '? Winter': 'Winter Olympics',
               '01 !.1': 'Gold.1',
               '02 !.1': 'Silver.1',
               '03 !.1': 'Bronze.1',
               '? Games': '# Games',
               '01 !.2': 'Gold.2',
               '02 !.2': 'Silver.2',
               '03 !.2': 'Bronze.2'}

olympic.rename(columns= col_names, inplace = True)
olympic.head(2)
```

```python
# we need to remove the code from the Country name
# create a function thas trim the country name 
def trim_country(str):
        # Find the index of the first occurrence of " ("
        index = str.find(' (')
        # trim the str till that index
        country = str[:index]
        return(country)

#trim_country('INIDA BHARAT (IND) (BHA) ')
```

```python
olympic['Country']= olympic['Country'].apply(trim_country)
olympic
```
