# Instruction ( For Data Cleaning ) - Find all Null Values in the dataset. 
- If there is any null value in any column, 
- then fill it with the mean of that column.

```python
#If there is any null value in any column
cars.isnull().sum()
```

```python
# Replace NaN values with the mean of numeric columns
num_col = cars.select_dtypes(include='number').columns
num_df = cars[num_col].fillna(cars[num_col].mean())
str_col = cars.select_dtypes(include='object').columns
str_df = cars[str_col].fillna('Not')
#mearginf both dataframe
final_cars = pd.concat([num_df,str_df],axis =1)
final_cars
```

```python
#check for clean data
final_cars.isnull().sum()
```
